## spaCy is an open-source library used for natural language processing tasks. It comes pre-installed in Google Colab but it is better to check which version is available before starting.

```python
# importing the spacy library
import spacy
```

```python
# creating the English nlp object
nlp = spacy.load('en_core_web_sm')
```

```python
# class type of the object
type(nlp)
```

##Tokenization

```python
# sample sentence
corpus = "The price of my playstation was $499 and it included a game disk."

# processing the text (tokenizing) to return a Doc object
doc = nlp(corpus)
```

```python
# printing the tokenized text
print([t.text for t in doc])
```

```python
# printing only the first tokenized text using index method
print(doc[0])
```

```python
# printing a span of tokens using the slicing method
print(doc[0:4])  # slicing a doc object returns a span object
```

```python
# using the 'i' attribute to represent index of each tokens in the doc object
print([(t.text, t.i) for t in doc])
```

```python
# spacy's tokenization doesn't destroy the original input so it can be viewed
print(doc.text)
```
